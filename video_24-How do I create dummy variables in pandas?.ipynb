```python
# tekrar orjinal datayƒ± √ßaƒüƒ±rƒ±yoruz
train = pd.read_csv("http://bit.ly/kaggletrain")
```

```python
#diƒüer get_dummies() metodudur. 
pd.get_dummies(train.Sex,prefix='Sex').iloc[:,1:]
```

```python
# ba≈üka birinde deneyelim.
train.Embarked.value_counts()
```

```python
# eƒüer n tane category varsa n-1 tane dummied variable olmak zorundadƒ±r. 
# budurumda 1den sonraki columnlarƒ± alƒ±yoruz.
# yeni bir df yapƒ±yoruz.
embarked_dummies = pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:,1:]
```

```python
# orjinal df le yer deƒüi≈ütiriyoruz.
train = pd.concat([train,embarked_dummies],axis=1)
```

```python
# yeni alanlarƒ±n eklendiƒüini g√∂r√ºr√ºz. 
train.head()
```

```python
# aynƒ±sƒ± sex alanƒ± i√ßinde yapalƒ±m.
train.Sex.value_counts()
```

```python
sex_dummies = pd.get_dummies(train.Sex, prefix='Sex').iloc[:,1:]
```

```python
# orjinal df le yer deƒüi≈ütiriyoruz.
train = pd.concat([train,sex_dummies],axis=1)
```

```python
# sex dummy variable larƒ± ekledik. 
train.head()
```

```python
# tekrar orjinal datayƒ± √ßaƒüƒ±rƒ±yoruz
train = pd.read_csv("http://bit.ly/kaggletrain")
```

```python
#orjinal alanlarƒ±n gittiƒüini yerine dummyied alanlarƒ±n geldiƒüini g√∂r√ºr√ºz.
# drop_first ise ilk dummy yi kaldƒ±rƒ±r. bu sayede manual i≈ü yapmamƒ±≈ü oluruz.
train = pd.get_dummies(train,columns=['Sex','Embarked'], drop_first=True) 
```
